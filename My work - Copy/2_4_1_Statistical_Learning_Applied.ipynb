- $v$. Use the `hist()` function to produce some histograms with differing numbers of bins for a few of the quantitative variables. You may find the command `par(mfrow=c(2,2))` useful: it will divide the print window into four regions so that four plots can be made simultaneously. Modifying the arguments to this function will divide the screen in other ways.

```python
def plot_hist_grid(df, figsize, layout, bins):
    ''' plot a grid of hist from all columns in the df
    df      -> a dateframe with only quantitative variables
    figsize -> the figure size
    layout  -> the shape of the grid
    bins    -> number of bins bars
    '''
    
    # Resize the chart, and have many hist plots in the shape of the layout
    # Set a larger figure size for subplots
    plt.figure(figsize=figsize) 

    for plot in range(df.shape[1]):
        # histogram on the layout point
        # 3 rows, 4 cols, subplot from 1 to the len of cols
        plt.subplot(layout[0], layout[1], plot+1) 
        sns.histplot(df.iloc[:,plot], bins=bins, color=np.random.rand(3));
```

```python
plot_hist_grid(college_v1.iloc[:, 1:10],
               figsize=[16,10],
               layout=[3,3],
               bins = 25)
```
